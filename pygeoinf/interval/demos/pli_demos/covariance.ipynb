```python
from pygeoinf.interval import (IntervalDomain, Lebesgue, InverseLaplacian, BoundaryConditions,
                               KLSampler)
```

```python
function_domain = IntervalDomain(0, 1)

M = Lebesgue(0, function_domain, basis=None)
bcs = BoundaryConditions('dirichlet')
L = InverseLaplacian(M, bcs, 1, method='spectral')
sampler = KLSampler(L, n_modes=1000)
```

```python
sampler.sample().plot()
```
