```python
from pygeoinf.interval.lebesgue_space import Lebesgue
from pygeoinf.interval.interval_domain import IntervalDomain
from pygeoinf.interval.boundary_conditions import BoundaryConditions
from pygeoinf.interval.operators import LaplacianOperator, LaplacianInverseOperator, GradientOperator
from pygeoinf.interval.functions import Function
import numpy as np
from matplotlib import pyplot as plt
```

```python
func_domain = IntervalDomain(0.0, 1.0)
bcs = BoundaryConditions('dirichlet', left=0.0, right=0.0)
l2space = Lebesgue(8, func_domain, basis='fourier')
laplacian = LaplacianOperator(l2space, bcs, method='finite_difference', dofs=10)
```

```python
f = Function(l2space, evaluate_callable=lambda x: np.exp(x))
bcs = BoundaryConditions('dirichlet', left=0.0, right=0.0)
lap = LaplacianOperator(l2space, bcs, method = 'finite_difference', dofs=100)
g = lap(f)
```

```python
x = func_domain.uniform_mesh(100)

plt.plot(x, f(x), label='f')
plt.plot(x, g(x), label='grad f')
plt.legend()
plt.show()
```
