```python
from pygeoinf.subsets import Ball, SublevelSet
from pygeoinf.subspaces import AffineSubspace
from pygeoinf.plot import plot_subset_oracle
from pygeoinf import EuclideanSpace, NonLinearForm
import numpy as np
```

```python
M = EuclideanSpace(2)
form = NonLinearForm(M, lambda x: np.sin(x[0]) + (np.cos(x[1]))**2)
my_set = SublevelSet(form, level=0.0)
# define some directions in the space
basis_vectors = [M.from_components(np.array([1, 0])), M.from_components(np.array([0, 1]))]
x_0 = M.from_components(np.array([0.0, 0.0]))
cutting_plane = AffineSubspace.from_tangent_basis(M, basis_vectors, x_0, orthonormalize=False)
```

```python
plot_subset_oracle(my_set, on_subspace=cutting_plane,
                   show_plot=True, bounds=[-15, 15, -15, 15],
                   grid_size=500)
```
